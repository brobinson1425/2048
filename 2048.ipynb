```python
import tkinter as tk
from tkinter import ttk
import numpy
import copy
import random
import math

#Necessary Features
#Title
#Game Over Check and display
#Play again
#Score Counter
#Button Slide blocks




root = tk.Tk()
MainFrame = ttk.Frame(root)
def ColorUpdate():
    global playSpaces
    for x in range(16):
        if playSpaces[x]["text"] == "0": #Gray
            playSpaces[x].config(background = "#909090", foreground = "#909090")
        elif playSpaces[x]["text"] == "2": 
            playSpaces[x].config(background = "#61b0ff", foreground = "black")
        elif playSpaces[x]["text"] == "4": 
            playSpaces[x].config(background = "#61caff", foreground = "black")
        elif playSpaces[x]["text"] == "8": 
            playSpaces[x].config(background = "#61e5ff", foreground = "black")
        elif playSpaces[x]["text"] == "16": 
            playSpaces[x].config(background = "#61ffff", foreground = "black")
        elif playSpaces[x]["text"] == "32": 
            playSpaces[x].config(background = "#61ffe5", foreground = "black")
        elif playSpaces[x]["text"] == "64": 
            playSpaces[x].config(background = "#61ffca", foreground = "black")
        elif playSpaces[x]["text"] == "128": 
            playSpaces[x].config(background = "#61ffb0", foreground = "black")
        elif playSpaces[x]["text"] == "256": 
            playSpaces[x].config(background = "#61ff7b", foreground = "black")
        elif playSpaces[x]["text"] == "512": 
            playSpaces[x].config(background = "#7bff61", foreground = "black")
        elif playSpaces[x]["text"] == "1024": 
            playSpaces[x].config(background = "#96ff61", foreground = "black")
        elif playSpaces[x]["text"] == "2048": 
            playSpaces[x].config(background = "#b0ff61", foreground = "black")
            

def AddValue():
    global playSpaces, gameOver, slide
    spaceOccupied = True
    tries = 0
    spacesLeft = False
    for x in range(15):
        if playSpaces[x]["text"] == "0":
            spacesLeft = True
    if spacesLeft and slide:        
        while spaceOccupied:
            if gameOver:
                break
            location = random.randint(0,15)
            if playSpaces[location]["text"] == "0":
                spaceOccupied = False
            tries = tries+1
        Odds = random.random()
        if Odds < 0.9:
            Value = 2
        else:
            Value = 4
        playSpaces[location].config(text = str(Value))
        #print(location)
    
def checkLose():
    global playSpaces, gameOver
    gameOver = True
    
    
    for x in range(0,16):
        if playSpaces[x]["text"] == "0":
            gameOver = False
    for i in range(0,4):
        if playSpaces[i*4]["text"] == playSpaces[i*4+1]["text"]:
            gameOver = False
        elif playSpaces[i*4+1]["text"] == playSpaces[i*4+2]["text"]:
            gameOver = False
        elif playSpaces[i*4+2]["text"] == playSpaces[i*4+3]["text"]:
            gameOver = False
    for j in range(0,4):
        if playSpaces[0*4+j]["text"] == playSpaces[1*4+j]["text"]:
            gameOver = False
        elif playSpaces[1*4+j]["text"] == playSpaces[2*4+j]["text"]:
            gameOver = False
        elif playSpaces[2*4+j]["text"] == playSpaces[3*4+j]["text"]:
            gameOver = False
    
    
    if gameOver:
        print("GAME OVER")
        EndGameLabel.grid()
        EndGameLabel.config(text = "Game Over", foreground = "#ff6161")
        
def checkWin():
    global playSpaces
    win = False
    for x in range(0,16):
        if playSpaces[x]["text"] == "2048":
            win = True
    if win:
        EndGameLabel.grid()
        EndGameLabel.config(text = "You Win!!", foreground = "#8eff61")

def upKeyPressed(event): #Functional
    global playSpaces, slide
    for trial in [1,2,3]:
        for x in range(0,12):
            if playSpaces[x]["text"] == "0":
                playSpaces[x]["text"] = playSpaces[x+4]["text"]
                playSpaces[x+4]["text"] = "0"

                if playSpaces[x]["text"] != "0":
                    slide = True
            
    
    for j in range(0,4):
        if playSpaces[0*4+j]["text"] == playSpaces[1*4+j]["text"] and playSpaces[0*4+j]["text"] != "0":
            playSpaces[0*4+j]["text"] = str(int(playSpaces[0*4+j]["text"])+int(playSpaces[1*4+j]["text"]))
            playSpaces[1*4+j]["text"] = "0"
            slide = True
            
        if playSpaces[1*4+j]["text"] == playSpaces[2*4+j]["text"] and playSpaces[1*4+j]["text"] != "0":
            playSpaces[1*4+j]["text"] = str(int(playSpaces[1*4+j]["text"])+int(playSpaces[2*4+j]["text"]))
            playSpaces[2*4+j]["text"] = "0"
            slide = True
            
        if playSpaces[2*4+j]["text"] == playSpaces[3*4+j]["text"] and playSpaces[2*4+j]["text"] != "0":
            playSpaces[2*4+j]["text"] = str(int(playSpaces[2*4+j]["text"])+int(playSpaces[3*4+j]["text"]))
            playSpaces[3*4+j]["text"] = "0"
            slide = True
            
    for x in reversed(range(0,12)):
        if playSpaces[x]["text"] == "0":
            playSpaces[x]["text"] = playSpaces[x+4]["text"]
            playSpaces[x+4]["text"] = "0"
            
            if playSpaces[x]["text"] != "0":
                slide = True
                
    checkWin()            
    AddValue()
    ColorUpdate()
    checkLose()
    slide = False
    
def leftKeyPressed(event): #To Be Fixed
    global playSpaces, slide
    for trial in [1,2,3]:
        for x in range(0,16):
            if playSpaces[x]["text"] == "0" and x%4 != 3:
                playSpaces[x]["text"] = playSpaces[x+1]["text"]
                playSpaces[x+1]["text"] = "0"

                if playSpaces[x]["text"] != "0":
                    slide = True
    
    for i in range(0,4):
        if playSpaces[i*4+0]["text"] == playSpaces[i*4+1]["text"] and playSpaces[i*4+0]["text"] != "0":
            playSpaces[i*4+0]["text"] = str(int(playSpaces[i*4+0]["text"])+int(playSpaces[i*4+1]["text"]))
            playSpaces[i*4+1]["text"] = "0"
            slide = True
            
        if playSpaces[i*4+1]["text"] == playSpaces[i*4+2]["text"] and playSpaces[i*4+1]["text"] != "0":
            playSpaces[i*4+1]["text"] = str(int(playSpaces[i*4+1]["text"])+int(playSpaces[i*4+2]["text"]))
            playSpaces[i*4+2]["text"] = "0"
            slide = True
            
        if playSpaces[i*4+2]["text"] == playSpaces[i*4+3]["text"] and playSpaces[i*4+2]["text"] != "0":
            playSpaces[i*4+2]["text"] = str(int(playSpaces[i*4+2]["text"])+int(playSpaces[i*4+3]["text"]))
            playSpaces[i*4+3]["text"] = "0"
            slide = True
            
    for x in reversed(range(0,16)):
        if playSpaces[x]["text"] == "0" and x%4 != 3:
            playSpaces[x]["text"] = playSpaces[x+1]["text"]
            playSpaces[x+1]["text"] = "0"
            
            if playSpaces[x]["text"] != "0":
                slide = True
    
    checkWin()
    AddValue()
    ColorUpdate()
    checkLose()
    slide = False
    
def downKeyPressed(event): #Completed
    global playSpaces, slide
    for trial in [1,2,3]:
        for x in reversed(range(4,16)):
            if playSpaces[x]["text"] == "0":
                playSpaces[x]["text"] = playSpaces[x-4]["text"]
                playSpaces[x-4]["text"] = "0"

                if playSpaces[x]["text"] != "0":
                    slide = True
            
    
    for j in range(0,4):
        
        if playSpaces[2*4+j]["text"] == playSpaces[3*4+j]["text"] and playSpaces[2*4+j]["text"] != "0":
            playSpaces[3*4+j]["text"] = str(int(playSpaces[2*4+j]["text"])+int(playSpaces[3*4+j]["text"]))
            playSpaces[2*4+j]["text"] = "0"
            slide = True
            
        if playSpaces[1*4+j]["text"] == playSpaces[2*4+j]["text"] and playSpaces[1*4+j]["text"] != "0":
            playSpaces[2*4+j]["text"] = str(int(playSpaces[1*4+j]["text"])+int(playSpaces[2*4+j]["text"]))
            playSpaces[1*4+j]["text"] = "0"
            slide = True
            
        if playSpaces[0*4+j]["text"] == playSpaces[1*4+j]["text"] and playSpaces[0*4+j]["text"] != "0":
            playSpaces[1*4+j]["text"] = str(int(playSpaces[0*4+j]["text"])+int(playSpaces[1*4+j]["text"]))
            playSpaces[0*4+j]["text"] = "0"
            slide = True
            
    for x in range(4,16):
        if playSpaces[x]["text"] == "0":
            playSpaces[x]["text"] = playSpaces[x-4]["text"]
            playSpaces[x-4]["text"] = "0"
            
            if playSpaces[x]["text"] != "0":
                slide = True
     
    checkWin()
    AddValue()
    ColorUpdate()
    checkLose()
    slide = False
    
    
    
def rightKeyPressed(event): #To Be Fixed
    global playSpaces, slide
    for trial in [1,2,3]:
        for x in reversed(range(0,16)):
            if playSpaces[x]["text"] == "0" and x%4 != 0:
                playSpaces[x]["text"] = playSpaces[x-1]["text"]
                playSpaces[x-1]["text"] = "0"

                if playSpaces[x]["text"] != "0":
                    slide = True
    
    for i in range(0,4):
        if playSpaces[i*4+2]["text"] == playSpaces[i*4+3]["text"] and playSpaces[i*4+3]["text"] != "0":
            playSpaces[i*4+3]["text"] = str(int(playSpaces[i*4+2]["text"])+int(playSpaces[i*4+3]["text"]))
            playSpaces[i*4+2]["text"] = "0"
            slide = True
            
        if playSpaces[i*4+1]["text"] == playSpaces[i*4+2]["text"] and playSpaces[i*4+2]["text"] != "0":
            playSpaces[i*4+2]["text"] = str(int(playSpaces[i*4+1]["text"])+int(playSpaces[i*4+2]["text"]))
            playSpaces[i*4+1]["text"] = "0"
            slide = True
            
        if playSpaces[i*4+0]["text"] == playSpaces[i*4+1]["text"] and playSpaces[i*4+1]["text"] != "0":
            playSpaces[i*4+1]["text"] = str(int(playSpaces[i*4+0]["text"])+int(playSpaces[i*4+1]["text"]))
            playSpaces[i*4+0]["text"] = "0"
            slide = True
            
        
            
    for x in range(0,16):
        if playSpaces[x]["text"] == "0" and x%4 != 0:
            playSpaces[x]["text"] = playSpaces[x-1]["text"]
            playSpaces[x-1]["text"] = "0"
            
            if playSpaces[x]["text"] != "0":
                slide = True
                
    checkWin()           
    AddValue()
    ColorUpdate()
    checkLose()
    slide = False


gameOver = False
slide = True

QuitButton = ttk.Button(MainFrame, text="Quit", command=root.destroy)
QuitButton.grid(row = 5, column=3)



gameTitle =  ttk.Label(MainFrame, foreground="Black", text="2048", anchor=tk.CENTER, font='-weight bold -size 20')
gameTitle.grid(row=0,column=0, columnspan=4, pady = 10, ipadx = 100)

style = ttk.Style(root)
style.configure("TFrame", background = "#707070")

playFrame = ttk.Frame(MainFrame, style="TFrame")
playFrame.grid(row = 1, column = 0, rowspan = 4, columnspan = 4)
#playFrame.config(background = "#707070")

playSpaces = []

for i in range(4):
    for j in range(4):
        playSpace = ttk.Label(playFrame, width=4, background='white', foreground="black", text="0", anchor=tk.CENTER, font='-weight bold -size 20')
        
        playSpace.grid(row=i, column=j, ipady = 15, padx = 5, pady = 5)
        playSpaces.append(playSpace)


#print(matrix)
EndGameLabel = ttk.Label(MainFrame, text="2048", anchor=tk.CENTER, font='-weight bold -size 20')
EndGameLabel.grid(row=2, column=0, columnspan=4, rowspan = 2, pady = 10, ipadx = 100)
EndGameLabel.grid_remove()
           
AddValue()
ColorUpdate() 
    
MainFrame.grid()
    
root.bind("w",upKeyPressed)
root.bind("a",leftKeyPressed)
root.bind("s",downKeyPressed)
root.bind("d",rightKeyPressed)

root.bind("<Up>",upKeyPressed)
root.bind("<Left>",leftKeyPressed)
root.bind("<Down>",downKeyPressed)
root.bind("<Right>",rightKeyPressed)


root.update()
screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()
window_width = root.winfo_width()
window_height = root.winfo_height()
#print(screen_width,screen_height)
root.geometry("+%d+%d" %((screen_width-window_width)/2,(screen_height-window_height)/2))
root.focus_force()
#def on_key_press(event):
    #print(event.keysym)
root.mainloop()
```
